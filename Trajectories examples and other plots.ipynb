```python
import numpy as np
import molecCollPackage as molec
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

# Examples of collisions

```python
# Examples of collisions
for T in [1, 10, 100, 500]:
    t, basin, x1out, y1out, z1out, x2out, y2out, z2out, x3out, y3out, z3out = molec.run_collision(T=T, t=(0, 10000), d_molec = 8.78, dt = 1, rinit1=60.0, theta = np.pi/2, atom="Li")[:]
    plt.figure(figsize=(7,7))
    plt.plot(y1out,x1out,'C2')
    plt.plot(y2out,x2out,'C3')
    plt.plot(y3out,x3out,'C4')
    plt.xlim(-20,20)
    plt.ylim(-20,20)
    plt.ylabel("a_0 ")
    plt.title("Collision energy = %s K, theta = pi/2" %(T))
    plt.show()
```

```python
# Examples of collision
for theta_frac in [0.5, 0.25, 0.33, 0.1]:
    theta = np.pi*theta_frac
    t, basin, x1out, y1out, z1out, x2out, y2out, z2out, x3out, y3out, z3out = molec.run_collision(T=10, theta = theta, t=(0, 10000), d_molec = 8.78, dt = 1, rinit1=60.0, atom="Li")[:]
    plt.figure(figsize=(7,7))
    plt.plot(y1out,x1out,'C2')
    plt.plot(y2out,x2out,'C3')
    plt.plot(y3out,x3out,'C4')
    plt.xlim(-20,20)
    plt.ylim(-20,20)
    plt.ylabel("a_0 ")
    plt.title("Collision energy = 10 K, theta = %s x Pi" %(theta_frac))
    plt.show()
```

# Lenard-Jones potential

```python
def exag_LJ(r):
     return 10 / r ** 12 - 500 / r ** 6
```

```python
x = np.linspace(0.5, 2, 1000)
y = exag_LJ(x)

fig = plt.figure()
plt.plot(x, y)
plt.title("Lenard-Jones potential (arbitrary units)")
plt.ylabel("Energy")
plt.xlabel("Distance")
plt.show()
pp = PdfPages('LJ.pdf')
pp.savefig(fig)
pp.close()
```
